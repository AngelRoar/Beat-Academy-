## Visualization with Pandas Plotting 

### Histogram

```python
data.columns
```

```python
data['Age'].plot.hist();
```

```python
data['Age'].plot.hist(cmap='rainbow');
```

```python
data['Fare'].plot.hist(cmap='magma');
```

### Box Plot

```python
data['Fare'].plot.box(cmap='viridis');
```

```python
data.Fare.plot.box(cmap='plasma');
```

### Kernel Density Estimate plot

```python
data.Fare.plot.kde(cmap='plasma');
```

```python
data.Age.plot.kde(cmap='rocket');
```

Subplot

```python
data.plot.box(subplots=True);
```

```python
# This chart is very hard to read, so we need to rework and give sizes for the chart 
```

```python
data.plot.box(figsize=(15,3),subplots=True);
```

```python
data.plot.hist(figsize=(3,13),subplots=True);
```

```python
data.plot.kde(figsize=(5,12),subplots=True);
```

### Line Chart

```python
data.plot.line(x='Age', y='Fare', cmap='mako')
```

```python
# This occures because th Age was not in order
```

```python
data.sort_values(by='Age').plot.line(x='Age', y='Fare', cmap='Wistia');
```

### Count plot

```python
data.Cabin.value_counts()
```

```python
data.Cabin.value_counts().plot.bar(figsize=(40,7));
```

```python
data.Age.value_counts().plot.bar(figsize=(17,4), cmap = 'Paired');
```

### Area chart

```python
data.plot.area(x='Age', y='Pclass', figsize = (10,6), cmap = 'rainbow');
```

### Horizontal bar chart

```python
data.nlargest(5, 'Fare').plot.barh(x = 'Embarked', y = 'Fare');
```

```python
data.sort_values(by='Fare')[:5].plot.barh(x='Age', y= 'Fare', figsize = (6,4), cmap = 'flag');
```

### Scatter plot

```python
data.plot.scatter(x='Age', y='Fare');
```
