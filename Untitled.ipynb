Convert the integers 0, 1, and -5 to booleans

```python
num_1 = 0
num_2 = 1
num_3 = -5
num_1 = bool(num_1)
num_2 = bool(num_2)
num_3 = bool(num_3)
print(num_1)
print(num_2)
print(num_3)
```

Convert the strings "Hello" and "" (empty string) to booleans.

```python
type_1 = "Hello"
type_2 = ""
type_1 = bool(type_1)
type_2 = bool(type_2)
print(type_1)
print(type_2)
```
